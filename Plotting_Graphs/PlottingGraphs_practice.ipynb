##  x^3 and x^2

```python
import matplotlib.pyplot as plt
import numpy as np

#x = np.array([1,2,3,4]) # Not like curve bcz step size is 1 which is very large
x = np.arange(0,5,0.1) # step size=0.1
y = x**3
y1 = x**2

plt.plot(x,y,'r', linewidth=3, label='x^3')
plt.plot(x,y1, color='black', marker='+', linewidth=3, label='x^2')
#plt.axis([0,10,0,150]) # x_stat, x_end, y_start, y_end
plt.text(3,27, '(3 , 27)', fontsize=12)
plt.grid()

plt.xlabel('X')
plt.ylabel('Y')
plt.title('X^2 and X^3')
plt.legend()
plt.show()
```

## Bubble Graph

```python
import matplotlib.pyplot as plt
import numpy as np

year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,18]
population = [100,120,180,250,300,370]

#plt.scatter(year,salary, s=100)
#plt.scatter(year,salary, s=population, c='r')
color_arr = np.arange(len(year))
plt.scatter(year,salary, s=population, c=color_arr, alpha=0.7, edgecolor='black', linewidth=3, marker='p') # marker='p' means pentagon

plt.text(year[0]+0.1, salary[0]+0.1, population[0])
plt.title('Bubble chart example')
plt.show()
```

## Pie Graph

```python
import matplotlib.pyplot as plt

sizes=[3,4,6,2]
labels = ['A','B','C','D']
colors = ['blue','purple','red','pink']
explode = [0.2,0.1,0,0] #to separate out some section of the graph

plt.pie(sizes, colors=colors, explode=explode, labels=labels, autopct='%.2f%%', counterclock=False, startangle=45)
plt.axis("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt

company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
plt.pie(sold,labels=company, autopct='%.1f')
plt.axis("equal")
plt.show()

total = sum(sold)
for i in range(6):
    p = (sold[i]*100)/total
    print(company[i], format(p,'.1f'))
```
