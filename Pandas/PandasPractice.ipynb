## Find the data of flower “Iris-virginica” type where petal-length > 1.5

## Replace NaN values

```python
import numpy as np

df = iris.copy()
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']

# add some nan values
df.iloc[2:4, 1:3] = np.nan
df.head()
```

```python
# Replace nan with mean value of their column
df.sw.fillna(df.sw.mean(), inplace=True)
df.pl.fillna(df.pl.mean(), inplace=True)

# We can also replace with mean value of particular type of flower
df.head()
```

## Handling String Values

```python
df = iris.copy()
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']

#add column
df['gender'] = 'female'

# make 1st 4 entries as male
df.iloc[:4, 5] = 'male'

df.head()
```

```python
# replace male by 0 and female by 1
def f(s):
    if s=='male':
        return 0
    else:
        return 1
    
df['sex'] = df.gender.apply(f)

df.drop('gender', axis=1, inplace=True)
df.head()
```

## Find and print the minimum, maximum and average value of the feature for each kind of flower

```python
import pandas as pd
import numpy as np
column_names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=column_names)

f1 = df[df['Species']=='Iris-setosa']
f2 = df[df['Species']=='Iris-versicolor']
f3 = df[df['Species']=='Iris-virginica']

f1_min = f1.min()
f1_max = f1.max()
f1_avg = f1.mean()
for i in f1_min[:-1]: 
    print(format(i,'.2f'),end=" ")
print('Iris-setosa')
for i in f1_max[:-1]:
    print(format(i,'.2f'),end=" ")
print('Iris-setosa')
for i in f1_avg:
    print(format(i,'.2f'),end=" ")
print('Iris-setosa')

# for f2
f2_min = f2.min()
f2_max = f2.max()
f2_avg = f2.mean()
for i in f2_min[:-1]:
    print(format(i,'.2f'),end=" ")
print('Iris-versicolor')
for i in f2_max[:-1]:
    print(format(i,'.2f'),end=" ")
print('Iris-versicolor')
for i in f2_avg:
    print(format(i,'.2f'),end=" ")
print('Iris-versicolor')

# for f3
f3_min = f3.min()
f3_max = f3.max()
f3_avg = f3.mean()
for i in f3_min[:-1]:
    print(format(i,'.2f'),end=" ")
print('Iris-virginica')
for i in f3_max[:-1]:
    print(format(i,'.2f'),end=" ")
print('Iris-virginica')
for i in f3_avg:
    print(format(i,'.2f'),end=" ")
print('Iris-virginica')
```
