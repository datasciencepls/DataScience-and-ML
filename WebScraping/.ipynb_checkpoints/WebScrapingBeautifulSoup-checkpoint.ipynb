### Print the difference between the number of descendants and the number of children of the html tag

```python
from bs4 import BeautifulSoup
data = BeautifulSoup(html,'html.parser')
l1 = len(list(data.html.children))
l2 = len(list(data.html.descendants))
print(l2-l1)
```

### Print the name of all the tags in different lines that have an id attribute.

```python
from bs4 import BeautifulSoup
data = BeautifulSoup(html,'html.parser')
li = data.find_all(True)
for i in li:
    if i.get('id',-1)!=-1:
        print(i.name)
```

```python
# 2nd way
from bs4 import BeautifulSoup
data = BeautifulSoup(html,'html.parser')
li = data.find_all(id = True)
for i in li:
    print(i.name)
```

### Print all content of the next siblings of the tag that have id as “li2”(in different lines)
### NOTE : Content includes the complete html of tag

```python
from bs4 import BeautifulSoup
data = BeautifulSoup(html,'html.parser')
li = data.find_all(id = 'li2')
ns = list(li[0].next_siblings) #li[0].next_siblings gives iterator so convert into list
for i in ns:
    print(i)
```

```python
# Way 2
from bs4 import BeautifulSoup
data = BeautifulSoup(html,'html.parser')
li = data.find(id = 'li2')
ns = li.next_siblings #li.next_siblings gives iterator so convert into list
for i in ns:
    print(i)
```

### Print content of all the parents of the title tag(linewise)
### NOTE : Content includes the whole HTML of the tag

```python
from bs4 import BeautifulSoup
data = BeautifulSoup(html,'html.parser')
parents = data.title.parents
for p in parents:
    print(p)
```

### Print the string which is present inside the isecond anchor tag using BeautifulSoup's next_element property

```python
from bs4 import BeautifulSoup
data = BeautifulSoup(html,'html.parser')
second_anchor = data.find_all('a')[1] # It return the second anchor tag
str = second_anchor.next_element # It gives the string inside 2nd anchor
print(str)
```

## Scraping website link : http://books.toscrape.com/

```python
from bs4 import BeautifulSoup
import requests

response = requests.get('http://books.toscrape.com/')
print(response)
```

```python
html_data = response.text # in form of string
data = BeautifulSoup(html_data,'html.parser')
print(data.prettify())
```

```python
print(data.header.a.string)
```

### Get title of 1st book

```python
b1 = data.find(class_='product_pod')
print(b1.h3.a['title']) # title is attribute 
```

### Hyperlinks of all books on first page

```python
books = data.find_all(class_='product_pod')
base_url = 'http://books.toscrape.com/'
for i in books:
    print(base_url + i.h3.a['href'])
```

### Print the title of all 20 books which are present on first page of http://books.toscrape.com/

```python
from bs4 import BeautifulSoup
import requests

response = requests.get('http://books.toscrape.com/')
data = BeautifulSoup(response.text, 'html.parser')

li = data.find_all(class_= 'product_pod')
for i in li:
    print(i.h3.a['title'])
```

### Print the name of all categories which are present http://books.toscrape.com/

```python
from bs4 import BeautifulSoup
import requests

response = requests.get('http://books.toscrape.com/')
data = BeautifulSoup(response.text, 'html.parser')

div = data.find(class_ = 'side_categories')
li = div.find_all('a') # 1st link is of Books. So leave that
for i in range(1,len(li)):
    print(li[i].string.strip())
```
