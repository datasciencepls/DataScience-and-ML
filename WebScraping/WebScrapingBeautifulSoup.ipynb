### For every book of 1st 10 pages store Title, Link, Price, Number of Copies in csv format

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html',
            'http://books.toscrape.com/catalogue/page-3.html',
            'http://books.toscrape.com/catalogue/page-4.html',
            'http://books.toscrape.com/catalogue/page-5.html',
            'http://books.toscrape.com/catalogue/page-6.html',
            'http://books.toscrape.com/catalogue/page-7.html',
            'http://books.toscrape.com/catalogue/page-8.html',
            'http://books.toscrape.com/catalogue/page-9.html',
            'http://books.toscrape.com/catalogue/page-10.html']

base_url = 'http://books.toscrape.com/catalogue/'
book_details = []
for i in allPages:
    page_url = i
    response = requests.get(page_url)
    data = BeautifulSoup(response.text,'html.parser')
    
    books = data.find_all(class_='product_pod')
    for b in books:
        book_b_url = base_url + b.h3.a['href']
        res = requests.get(book_b_url)
        b_data = BeautifulSoup(res.text, 'html.parser')
        
        title = b_data.h1.string
        #print(b.prettify())
        #print(book_b_url)
        price = b_data.find(class_ = 'price_color')
        qty = b_data.find(class_ = 'instock availability')
        qty = qty.contents[-1].strip()
        
        # To gt exact quantity and price we use regular expressions
        
        
        quantity = int(re.search('\d+',qty).group())
        pr = float(re.search('[\d.]+',str(price)).group())
        
        book_details.append([title, book_b_url, pr, quantity])

df = pd.DataFrame(book_details, columns = ['Title', 'Link', 'Price', 'Quantity In Stock'])
```

```python
df.to_csv('books.csv',index=False) # Store in CSV format
```
